```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'DeJavu Serif'
plt.rcParams['font.serif'] = ['Times New Roman']
```

```python
from sklearn.datasets import make_moons

# Veri seti
X, y = make_moons(n_samples=1000, noise=0.05, random_state=42)

# Sınıf 0
plt.scatter(X[y==0, 0], X[y==0, 1], color='blue', label='0', alpha=1)

# Sınıf 1
plt.scatter(X[y==1, 0], X[y==1, 1], color='red', label='1', alpha=1)

# Legend (açıklama kutusu)
plt.legend(loc='upper right')

# Başlık ve eksenler
#plt.title("Moon Veri Kümesi")
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid(False)

#plt.tight_layout()
# PDF olarak kaydet
#plt.savefig("moon.pdf", format='pdf')

plt.show()
```

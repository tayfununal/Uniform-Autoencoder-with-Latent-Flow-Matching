```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'DeJavu Serif'
plt.rcParams['font.serif'] = ['Times New Roman']
```

```python
# Veriyi yükle
digits = load_digits()

# İlk 16 görüntüyü al
images = digits.images[:16]
labels = digits.target[:16]

# 5x5 plot alanı oluştur
fig, axes = plt.subplots(4, 4, figsize=(8, 9), sharex=True, sharey=True,
                        subplot_kw={'xticks': [], 'yticks': []})

#fig.suptitle("Digits Veri Kümesi", fontsize=16)
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='binary')
    ax.set_title(f"Etiket: {labels[i]}")
    #ax.axis('off')  # Eksenleri gizle

# PDF olarak kaydet
plt.savefig('digits.pdf', format='pdf', bbox_inches='tight', pad_inches=0.1)
plt.show()
```

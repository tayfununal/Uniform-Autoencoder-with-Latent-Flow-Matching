```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'DeJavu Serif'
plt.rcParams['font.serif'] = ['Times New Roman']
```

```python
# Load the data
digits = load_digits()

# Take the first 16 images
images = digits.images[:16]
labels = digits.target[:16]

# Create a 5Ã—5 plot area
fig, axes = plt.subplots(4, 4, figsize=(8, 9), sharex=True, sharey=True,
                        subplot_kw={'xticks': [], 'yticks': []})


for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='binary')
    ax.set_title(f"Label: {labels[i]}")
    #ax.axis('off')

# PDF olarak kaydet
plt.savefig('digits.pdf', format='pdf', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
